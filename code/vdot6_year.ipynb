```python
import pandas as pd  
import numpy as np
import requests
from io import BytesIO
```

```python
# Use the raw GitHub URL
GITHUB_RAW_URL = "https://raw.githubusercontent.com/wkolichney/datathon2025_WeMightMaybeSuckAtBasicAlgorithms/main/data/vdot_rail_not_light.xlsx"

# Read Excel file from GitHub
response = requests.get(GITHUB_RAW_URL)
excel_data = BytesIO(response.content)
df = pd.read_excel(excel_data)

# Your existing code continues here
df = df[df['Road System'] == 'Rail']
df.head()






```

```python
#change columns that are listed in thousands to actual $
monetary_columns = ['Estimate', 'Previous', 'FY25', 'FY26-30', 'Balance']
for col in monetary_columns:
    df[col] = df[col] * 1000


print("Values now in actual dollars:")
df.head()

```

```python
#project estimated cost
total_estimate = df['Estimate'].sum()

print(f"\nTotal estimated cost for all rail projects: ${total_estimate:,.2f}")
```
